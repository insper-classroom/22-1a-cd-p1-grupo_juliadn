```python
#limpa as pontuações do texto
import re 


def cleanup(text):
    """
        Função de limpeza muito simples que troca alguns sinais básicos por espaços
    """
    #import string
    punctuation = '[!-.:?;"()''\/@*%$<>,[}{&]' # Note que os sinais [] são delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text.lower())
    return text_subbed
    
def laplace(palavra, tabela_abs, portugues):
    if palavra not in tabela_abs:
        prob_palavra_dado_categoria = 1 / ( len(portugues) + sum(tabela_abs) )
    else: 
        prob_palavra_dado_categoria = (tabela_abs[palavra] + 1 ) / ( len(portugues) + sum(tabela_abs) )
    
    return prob_palavra_dado_categoria


#stopwords função para remover as palavras sem importância
def stopwords(listapalavras):
    stopwords = nltk.corpus.stopwords.words('portuguese')
    filtrada = []
    for palavra in listapalavras:
        if palavra not in stopwords:
            filtrada.append(palavra)
    return filtrada
```

# La place

```python
test.fillna('', inplace=True)



listaclassi = []
for noticia in test.Titulo:
    noticia = cleanup(noticia.lower())
    palavras_limpas = stopwords(noticia.split())
    
    prob_neg_dado_noticia = 1
    prob_p_dado_noticia = 1
    prob_n_dado_noticia = 1
    
    for palavra in palavras_limpas:
        prob_neg_dado_noticia = prob_neg_dado_noticia * laplace(palavra, tabela_negativa, tabela_total)
        
        prob_n_dado_noticia = prob_n_dado_noticia * laplace(palavra, tabela_neutra, tabela_total)

        prob_p_dado_noticia = prob_p_dado_noticia * laplace(palavra, tabela_positiva, tabela_total)
        
    prob_neg_dado_noticia *= prob_priorinegativa
    
    prob_n_dado_noticia *= prob_priorineutra
    
    prob_p_dado_noticia *= prob_prioripositivo
    
    
    if prob_neg_dado_noticia > prob_n_dado_noticia and prob_neg_dado_noticia > prob_p_dado_noticia:
        listaclassi.append('Negativo')
    elif prob_n_dado_noticia > prob_neg_dado_noticia and prob_n_dado_noticia > prob_p_dado_noticia:
        listaclassi.append('Neutro')
    elif prob_p_dado_noticia > prob_neg_dado_noticia and prob_p_dado_noticia > prob_n_dado_noticia:
        listaclassi.append('Positivo')

test['Classificado'] = listaclassi

tab= pd.crosstab( test['Target'],test['Classificado'], normalize='index')
tab * 100
```

```python
verdadeiros_negativos = tab.iloc[0,0]*100
falsos_negativos = (tab.iloc[0,1] + tab.iloc[0,2])*100
verdadeiros_neutros = (tab.iloc[1,1])*100
falsos_neutros = (tab.iloc[1,0] + tab.iloc[1,2])*100
verdadeiros_positivos = (tab.iloc[2,2])*100
falsos_positivos = (tab.iloc[2,0] + tab.iloc[2,1])*100

eficacia = ( verdadeiros_negativos * verdadeiros_neutros * verdadeiros_positivos ) / 10000


print('A porcentagem de verdadeiros positivos é: {:.2f}%\n'.format(verdadeiros_positivos))
print('A porcentagem de falsos positivos é: {:.2f}%\n'.format(falsos_positivos))
print('A porcentagem de verdadeiros negativos é: {:.2f}%\n'.format(verdadeiros_negativos))
print('A porcentagem de falsos negativos é: {:.2f}%\n'.format(falsos_negativos))
print('A porcentagem de verdadeiros neutros é: {:.2f}%\n'.format(verdadeiros_neutros))
print('A porcentagem de falsos neutros é: {:.2f}%\n'.format(falsos_neutros))
print('Portanto, a eficácia do classificador é de: {:.2f}%\n'.format(eficacia))
```

```python
perf = []

for i in range(100):
    planilhas = ['Treinamento', 'Teste']
    dados = pd.DataFrame()
    
    for i in planilhas:
        t = pd.read_excel(filename, sheet_name=i)
        #concatena os dois dataframes
        dados = pd.concat([dados, t], axis=0, ignore_index=True)
        
    #função sklearn que irá redistribuir elas em treinamento e teste
    train,test = sklearn.model_selection.train_test_split(dados, train_size=0.6)

    #limpando as duas bases de dados
    
    #planilha de treino
    train.fillna("", inplace=True) #preenche colunas vazias para não dar erro 
    train['Titulo'] = train['Titulo'].apply(cleanup)

    #planilha de teste
    test.fillna("", inplace=True) #preenche colunas vazias para não dar erro
    test['Titulo'] = test['Titulo'].apply(cleanup)

    #dividindo de acordo com os targets
    positivos = train.loc[train['Target'] == 2]
    negativos = train.loc[train['Target'] == 0]
    neutros = train.loc[train['Target'] == 1]

    #transformando em palavras filtradas
    titulos_positivos = list(positivos['Titulo'])
    titulos_negativos = list(negativos['Titulo'])
    titulos_neutros = list(neutros['Titulo'])
    

    palavrastitulos_positivos = stopwords(' '.join(titulos_positivos).split())
    palavrastitulos_negativos = stopwords(' '.join(titulos_negativos).split())
    palavrastitulos_neutros = stopwords(' '.join(titulos_neutros).split())

    titulos_totais = palavrastitulos_negativos + palavrastitulos_positivos + palavrastitulos_neutros
    
    #series
    seriepositiva = pd.Series(palavrastitulos_positivos)
    serienegativa = pd.Series(palavrastitulos_negativos)
    serieneutra = pd.Series(palavrastitulos_neutros)
    serie_total = pd.Series(titulos_totais)
    
    
    #feq absoluta
    titulos_positivosabsoluta = seriepositiva.value_counts()
    titulos_negativosabsoluta = serienegativa.value_counts()
    titulos_neutrosabsoluta = serieneutra.value_counts()
    titulos_totaisabsoluta = serie_total.value_counts()

    #prob priori
    probp_positiva = len(palavrastitulos_positivos) / len(titulos_totais)
    probp_negativa = len(palavrastitulos_negativos) / len(titulos_totais)
    probp_neutras = len(palavrastitulos_neutros) / len(titulos_totais)

        
    performance = []

    for noticia_titulo in test.Titulo:

        noticia_titulo = cleanup(noticia_titulo)
        palavras_titulo = stopwords(noticia_titulo.split())

        probFrase_dado_PositivaT = 1
        probFrase_dado_NegativaT = 1
        probFrase_dado_NeutraT = 1

        for p in palavras_titulo:

            probFrase_dado_PositivaT = probFrase_dado_PositivaT * laplace(p, titulos_positivosabsoluta, titulos_totaisabsoluta)
            probFrase_dado_NegativaT = probFrase_dado_NegativaT * laplace(p, titulos_negativosabsoluta, titulos_totaisabsoluta)
            probFrase_dado_NeutraT = probFrase_dado_NeutraT * laplace(p, titulos_neutrosabsoluta, titulos_totaisabsoluta)


        probFrase_dado_PositivaT *= probp_positiva
        probFrase_dado_NegativaT *= probp_negativa
        probFrase_dado_NeutraT *= probp_neutras

        #adicionando na lista de classificação os targets de acordo com as probabilidades dos títulos
        if probFrase_dado_PositivaT > probFrase_dado_NegativaT and probFrase_dado_PositivaT > probFrase_dado_NeutraT:
            performance.append('Positiva')
        elif probFrase_dado_NegativaT > probFrase_dado_PositivaT and probFrase_dado_NegativaT > probFrase_dado_NeutraT:
            performance.append('Negativa')
        elif probFrase_dado_NeutraT > probFrase_dado_PositivaT and probFrase_dado_NeutraT > probFrase_dado_NegativaT:
            performance.append('Neutra')

    target_t = pd.Series(performance)
    test['Classifica_Título'] = target_t

    t_crosstab = pd.crosstab(test['Target'], test['Classifica_Título'], normalize='index')


    neg_verdadeiros = t_crosstab.iloc[0, 0] * 100
    neu_verdadeiros = t_crosstab.iloc[1, 1] * 100
    pos_verdadeiros = t_crosstab.iloc[2, 2] * 100
    
    eficacia_t = neg_verdadeiros * neu_verdadeiros * pos_verdadeiros / 100e2
    perf.append(eficacia_t)
    
perf = pd.Series(perf)
```
